3) O dataset Iris não possui valores ausentes. A base escolhida também pode conter todos os descritores por cada amostra. Use um dataset e sobre as amostras impute valores ausentes usando a média, mediana ou outras medidas.

Para verificar se há valores ausentes por coluna:

```python
print(data.isnull().sum())
```

Verificando a quantidade total de null:

```python
print(data.isnull().sum().sum())
```

```python
data.loc[data.isnull().any(axis=1)]
```

Simulando a existência de valores ausentes:

```python
data_copia = data.copy()
```

```python
cols =data.select_dtypes(include=['float64', 'int64']).columns
```

```python
for col in cols:
    faltando = np.random.choice(data.index, size=int(len(data_copia)*0.1), replace=False)
    data_copia.loc[faltando, col] = np.nan
```

Verificando se agora há valores ausentes:

```python
print(data_copia.isnull().sum().sum())
```

Métodos de imputação:

Por média:

```python
data_media = data_copia.copy()
for col in data_copia.select_dtypes(include=['float64', 'int64']).columns:
    data_media[col].fillna(data_media[col].mean(), inplace=True)
```

```python
print(data_media.isnull().sum().sum())
```

Por mediana:

```python
data_mediana = data_copia.copy()
for col in data_copia.select_dtypes(include=['float64', 'int64']).columns:
    data_mediana[col].fillna(data_mediana[col].median(), inplace=True)
```

```python
print(data_mediana.isnull().sum().sum())
```

Por moda:

```python
data_moda = data_copia.copy()
for col in data_copia.select_dtypes(include=['float64', 'int64']).columns:
    data_moda[col].fillna(data_moda[col].mode()[0], inplace=True)
```

```python
print(data_moda.isnull().sum().sum())
```

5) Compare os resultados obtidos usando um gráfico box plot. Que atributos possuem distribuições mais simétricas? Há a presença de valores extremos, denominados de outliers, para os atributos?

```python
fig, axes = plt.subplots(3, 4, figsize=(16, 12))
fig.suptitle('Histogramas dos Atributos Numéricos', fontsize=16)    

for i, col in enumerate(data.select_dtypes(include=['float64', 'int64']).columns):
    axes[i // 4, i % 4].boxplot(data[col], patch_artist=True, 
                              boxprops=dict(facecolor='lightblue', alpha=0.7))
    axes[i//4, i % 4].set_title(col)
    axes[i//4, i % 4].grid(True, alpha = 0.3)
plt.tight_layout()
plt.show()
```

De acordo com o gráfico boxplot, os atributos mais simétricos são o Álcool seguido pelo atributo de Qualidade. Além disso, existem muitos outliers, principalmente em residual sugar e em chlorides.

6) No dataset escolhido selecione intervalos de amostragem (10%, 30% e 50% - Progressiva) e faça análises sobre os dados em relação as métricas estatísticas. O que você pode observar sobre os subconjuntos em relação as métricas estatísticas com a nova composição dos dados?

```python
plt.rcdefaults()  

amostras ={
    '10%': data.sample(frac=0.1, random_state=42),
    '30%': data.sample(frac=0.3, random_state=42),
    '50%': data.sample(frac=0.5, random_state=42),
    '100%': data
}

colunas = data.select_dtypes(include=['float64', 'int64']).columns[:4]

for col in colunas:
    for size, amostra in amostras.items():
        media = amostra[col].mean()
        mediana = amostra[col].median()
        desvio = amostra[col].std()
        mininmo = amostra[col].min()
        maximo = amostra[col].max()
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Comparação de amostras', fontsize=16)

for i, col in enumerate(colunas):
    linha = i // 2
    coluna_indx = i % 2
    porcentagens = ['10%', '30%', '50%', '100%']
    medias = [amostras[size][col].mean() for size in porcentagens]
    desvios = [amostras[size][col].std() for size in porcentagens]
    
    axes[linha, coluna_indx].bar(porcentagens, medias, yerr=desvios, capsize=5, 
                          color=['lightcoral', 'gold', 'lightblue', 'lightgreen'],
                          alpha=0.8)
    axes[linha, coluna_indx].set_title(col)
    axes[linha, coluna_indx].set_ylabel('Valor')
    axes[linha, coluna_indx].grid(True, alpha=0.3)
    axes[linha, coluna_indx].set_facecolor('white')

plt.show()
```

É possivel verificar que a média é mais estável que o desvio padrão em amostras pequenas. 

*Fixed Acidity:*
Médias muito consistentes entre todas as amostras (~8.0)
Barras de erro pequenas = baixa variabilidade
Conclusão: Métrica muito estável, mesmo com 10% da amostra

*Volatile Acidity:*
Também estável (~0.5)
Amostra de 50% tem um pouquinho mais de variação
Conclusão: Bem representada em todas as amostras

*Citric Acid:*
Média com consistência entre amostras de maior porcentagem (~0.27)
Barras de erro quase iguais
Conclusão: Amostras com pouca variância

*Residual Sugar:*
Média estável (~2.5)
Maior variabilidade (barras de erro maiores)
Conclusão: Atributo com mais dispersão natural
