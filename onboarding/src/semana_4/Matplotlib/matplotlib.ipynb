## **Matplolib**

Importando:

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Carregando arquivo .csv:

```python
data = pd.read_csv("/UFU/IC/onboarding/src/semana_4/classification_results_trial_0001.csv", sep=',')
print(data.head())
```

1. Crie um gráfico de barras mostrando a contagem de imagens para real_class (quantas "benigno" e "maligno" são na realidade).


```python
plt.figure(figsize=(8, 5))
data['real_class'].value_counts().plot(kind='bar')
plt.title('Distribuição das Classes Reais')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.show()
```

2. Crie outro gráfico de barras mostrando a contagem de imagens para predicted_class (quantas "benigno" e "maligno" o modelo previu).


```python
plt.figure(figsize=(8, 5))
data['predicted_class'].value_counts().plot(kind='bar')
plt.title('Distribuição das Classes Preditas')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.show()
```

3. Crie um histograma para a coluna prob_benign (probabilidade de ser "benigno").

```python
plt.figure(figsize=(8, 5))
plt.hist(data['prob_benign'], edgecolor='black')
plt.title('Distribuição das Probabilidades Benignas')
plt.xlabel('Probabilidade Benigna')
plt.ylabel('Frequência')
plt.show()
```

4. Crie um histograma para a coluna prob_malign (probabilidade de ser "maligno").

```python
plt.figure(figsize=(8, 5))
plt.hist(data['prob_malign'], edgecolor='black')
plt.title('Distribuição das Probabilidades Malignas')
plt.xlabel('Probabilidade Maligna')
plt.ylabel('Frequência')
plt.show()
```

5. Crie um gráfico de dispersão (scatter plot) onde o eixo X pode ser prob_benign e o eixo Y pode ser prob_malign. Pinte os pontos de forma diferente com base se o modelo acertou ou errou a predição da imagem. Isso pode ajudar a visualizar se os erros ocorrem em regiões de menor confiança (probabilidades próximas de 0.5).

```python
plt.figure(figsize=(8, 8))

acertos = data[data['real_class'] == data['predicted_class']]
erros = data[data['real_class'] != data['predicted_class']]

plt.scatter(acertos['prob_benign'], acertos['prob_malign'], c='green', label='Acertos')
plt.scatter(erros['prob_benign'], erros['prob_malign'], c='red', label='Erros')

plt.axhline(0.5, color='gray', linestyle='--')
plt.axvline(0.5, color='gray', linestyle='--')

plt.xlabel('Probabilidade Benigna')
plt.ylabel('Probabilidade Maligna') 
plt.title('Acertos (Verde) vs Erros (Vermelho)')
plt.legend()
plt.show()
```

6. Qual tipo de erro é mais comum (Falso Positivo ou Falso Negativo)? Crie um gráfico que ajude a visualizar isso. Dada a natureza dos dados médicos, qual tipo de erro seria mais preocupante e por quê?

```python
fp = len(data[(data['real_class'] == 'benign') & (data['predicted_class'] == 'malign')])  
fn = len(data[(data['real_class'] == 'malign') & (data['predicted_class'] == 'benign')])

plt.figure(figsize=(6, 6))
plt.bar(['Falsos Positivos', 'Falsos Negativos'], [fp, fn], color=['orange', 'purple'])
plt.title('Falsos Positivos vs Falsos Negativos')
plt.ylabel('Quantidade')

plt.show()
```
